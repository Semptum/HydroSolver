```julia
using Revise
using HydroSolver
using LinearAlgebra
using StaticArrays
using Statistics
using GLMakie,Makie
using IterativeSolvers
```

```julia
Point = SVector{2,Float64}
```

```julia
L_x = 2
L_y = 2

L_slot = 0.5
L_coflow = 0.5

U_slot = 1e3
T_slot = 300
U_coflow = 0.2e3 
T_coflow = 300

ν = 15
ρ = 1.1614e-9
```

```julia
N_x = 50
N_y = 50
x_range=range(0,stop=L_x,length=N_x)
y_range=range(0,stop=L_y,length=N_y)
mesh=HydroSolver.rectangle_cartesian_mesher(x_range,y_range);
N_faces = length(mesh.faces)
N_cells = length(mesh.cells)
```

```julia
U = zeros(Point, N_cells)
gradP = zeros(Point, N_cells);
```

```julia
BC_U = Vector{HydroSolver.BoundaryConditions{Point}}(undef,length(mesh.fCenters))
BC_P = Vector{HydroSolver.BoundaryConditions{Float64}}(undef,length(mesh.fCenters))

left_border = [-1 in v && mesh.fCenters[i][1] < mean(x_range[1:2]) for (i,v) in enumerate(mesh.faces)]
bottom_border = [-1 in v && mesh.fCenters[i][2] < mean(y_range[1:2]) for (i,v) in enumerate(mesh.faces)]
top_border = [-1 in v && mesh.fCenters[i][2] > mean(y_range[end-1:end]) for (i,v) in enumerate(mesh.faces)]

bottom_slot = [0 <= i[1] <= L_slot for i in mesh.fCenters] .* bottom_border
top_slot = [0 <= i[1] <= L_slot for i in mesh.fCenters] .* top_border
bottom_coflow = [L_slot < i[1] <= L_slot+L_coflow for i in mesh.fCenters] .* bottom_border
top_coflow = [L_slot < i[1] <= L_slot+L_coflow for i in mesh.fCenters] .* top_border


for i in 1:N_faces
    BC_U[i] = HydroSolver.None{Point}()
    BC_P[i] = HydroSolver.None{Float64}()
    if left_border[i]
        BC_U[i] = HydroSolver.Neumann{Point}(zero(Point))
        BC_P[i] = HydroSolver.Neumann{Float64}(zero(Float64))
    end
    if top_border[i] || bottom_border[i]
        BC_U[i] = HydroSolver.Dirichlet{Point}(zero(Point))
        BC_P[i] = HydroSolver.Neumann{Float64}(zero(Float64))
    end
    if bottom_slot[i]
        BC_U[i] = HydroSolver.Dirichlet{Point}(Point([0,U_slot]))
    end
    if top_slot[i]
        BC_U[i] = HydroSolver.Dirichlet{Point}(Point([0,-U_slot]))
    end
    if bottom_coflow[i]
        BC_U[i] = HydroSolver.Dirichlet{Point}(Point([0,U_coflow]))
    end
    if top_coflow[i]
        BC_U[i] = HydroSolver.Dirichlet{Point}(Point([0,-U_coflow]))
    end

end
```

```julia
Diffus = HydroSolver.diffusion(mesh, HydroSolver.Laplacian{:NoCorrection});
```

```julia
function SIMPLE_iter(mesh, U, BC_U, BC_P)
    Convec = HydroSolver.convection(mesh, U, BC_U, HydroSolver.Interpolation{:Upwind})
    trM = Convec - ν * Diffus
    M = trM.A
    A = Diagonal(M)
    iA = inv(A)
    interp = HydroSolver.interpolate(mesh,U,HydroSolver.Interpolation{:Upwind})
    interp_P = HydroSolver.apply_boundary_conditions(mesh,interp,BC_P)
    Sum=HydroSolver.summation(mesh)
    
    GradP = [zeros(length(mesh.cells)) for i in 1:2]
    nU = [zeros(length(mesh.cells)) for i in 1:2]
    for axis in 1:2
        u = [i[axis] for i in U]
        H = (A-M)*u
        m  = [i[axis] for i in trM.B]
        fA = [i[axis] for i in mesh.fAVecs]
        D_u = Sum*Diagonal(fA)*interp
        d_p = [i for i in Sum*Diagonal(fA)*interp_P.B] / ρ
        LHS = (D_u*iA+1e-8I)
        RHS = Vector(D_u*iA*(H-m-d_p))
        GradP[axis] .= LHS\RHS
        nU[axis] .= iA*H - iA*GradP[axis] # M\(GradP[axis]-m) #
    end
    G_P = [Point([ GradP[1][i] , GradP[2][i] ]) for i in 1:length(mesh.cells)]
    newU = [Point([ nU[1][i] , nU[2][i] ]) for i in 1:length(mesh.cells)]
    return G_P
end
```

```julia
U = [Point([0.0,0.0]) for i in 1:N_cells]
dt=1e-5
for i in 1:100
    Convec = HydroSolver.convection(mesh, U, BC_U, HydroSolver.Interpolation{:Upwind})
    trM = Convec - ν * Diffus
    U -= dt* trM*U
end
for i in 1:0
    U = fractional_step(mesh, U, BC_U, BC_P)
end

```

```julia
function fractional_step(mesh, U, BC_U, BC_P; dt=1e-6)
    Convec = HydroSolver.convection(mesh, U, BC_U, HydroSolver.Interpolation{:Upwind})
    interp = HydroSolver.interpolate(mesh,U,HydroSolver.Interpolation{:Upwind})
    Sum=HydroSolver.summation(mesh)
    U_star = U - dt*Convec*U
    U_star_star = U_star + dt* ν*Diffus*U_star
    RHS = Sum*Diagonal(mesh.fAVecs)*interp*U_star_star .* ρ/dt
    P = Diffus \ RHS
    GradP = [zeros(length(mesh.cells)) for i in 1:2]
    GradP[1] = Sum*Diagonal([i[1] for i in mesh.fAVecs])*interp*P
    GradP[2] = Sum*Diagonal([i[1] for i in mesh.fAVecs])*interp*P
    gradP = [Point([ GradP[1][i] , GradP[2][i] ]) for i in 1:length(mesh.cells)]
    return P
    new_U = U_star_star - dt/ρ * gradP
    return new_U
end
```
