```julia
using Revise
using HydroSolver
using LinearAlgebra
using StaticArrays
using Statistics
using GLMakie,Makie
using IterativeSolvers
using Unitful
using ProfileView
using JLD2
```

On spécifie le type spatial du mesh: des points 2D.

```julia
Point = SVector{2,Float64}
```

```julia
L_x = inrefunit(2u"mm")
L_y = inrefunit(2u"mm")

L_slot = inrefunit(0.5u"mm")
L_coflow = inrefunit(0.5u"mm")

U_slot = inrefunit(1u"m/s")
T_slot = inrefunit(300u"K")
U_coflow = inrefunit(0.2u"m/s")
T_coflow = inrefunit(300u"K")

ν = inrefunit(15e-6u"m^2/s")
ρ = inrefunit(1.1614u"kg/m^3");
```

```julia
sigmoid(x)=1/(1+exp(-x*5))
x_range=sigmoid.(0.3*tan.(((0:0.5e-2:0.5).-1/2)*pi*0.8))
x_range.-=first(x_range)
x_range*=L_y/maximum(x_range)
y_range=sigmoid.(0.3*tan.(((0:1e-2:1).-1/2)*pi*0.8))
y_range.-=first(y_range)
y_range*=L_y/maximum(y_range);
```

```julia
N_x = 100
N_y = 100
x_range=range(0,stop=L_x,length=N_x)
y_range=range(0,stop=L_y,length=N_y)
mesh=HydroSolver.rectangle_cartesian_mesher(x_range,y_range);
N_faces = length(mesh.faces)
N_cells = length(mesh.cells);
plot_faces,plot_vertices =  HydroSolver.mesh_to_polygons(mesh);
```

```julia
U = zeros(Point, N_cells)
P = zeros(Float64, N_cells);
```

```julia
BC_U_x = Vector{HydroSolver.BoundaryConditions{Float64}}(undef,length(mesh.fCenters))
BC_U_y = Vector{HydroSolver.BoundaryConditions{Float64}}(undef,length(mesh.fCenters))
BC_P = Vector{HydroSolver.BoundaryConditions{Float64}}(undef,length(mesh.fCenters))

left_border = [-1 in v && mesh.fCenters[i][1] < mean(x_range[1:2]) for (i,v) in enumerate(mesh.faces)]
right_border = [-1 in v && mesh.fCenters[i][1] > mean(x_range[end-1:end]) for (i,v) in enumerate(mesh.faces)]
bottom_border = [-1 in v && mesh.fCenters[i][2] < mean(y_range[1:2]) for (i,v) in enumerate(mesh.faces)]
top_border = [-1 in v && mesh.fCenters[i][2] > mean(y_range[end-1:end]) for (i,v) in enumerate(mesh.faces)]

bottom_slot = [0 <= i[1] <= L_slot for i in mesh.fCenters] .* bottom_border
top_slot = [0 <= i[1] <= L_slot for i in mesh.fCenters] .* top_border
bottom_coflow = [L_slot < i[1] <= L_slot+L_coflow for i in mesh.fCenters] .* bottom_border
top_coflow = [L_slot < i[1] <= L_slot+L_coflow for i in mesh.fCenters] .* top_border


for i in 1:N_faces
    BC_U_x[i] = HydroSolver.None{Float64}()
    BC_U_y[i] = HydroSolver.None{Float64}()
    BC_P[i] = HydroSolver.None{Float64}()
    if left_border[i]
        BC_U_x[i] = HydroSolver.Dirichlet{Float64}(0)
        BC_P[i] = HydroSolver.Neumann{Float64}(0)
    end
    if top_border[i] || bottom_border[i]
        BC_U_x[i] = HydroSolver.Dirichlet{Float64}(0)
        BC_U_y[i] = HydroSolver.Dirichlet{Float64}(0)
        BC_P[i] = HydroSolver.Neumann{Float64}(zero(Float64))
    end
    if bottom_slot[i]
        BC_U_x[i] = HydroSolver.Dirichlet{Float64}(0)
        BC_U_y[i] = HydroSolver.Dirichlet{Float64}(U_slot)
        BC_P[i] = HydroSolver.Dirichlet{Float64}(0)
    end
    if top_slot[i]
        BC_U_x[i] = HydroSolver.Dirichlet{Float64}(0)
        BC_U_y[i] = HydroSolver.Dirichlet{Float64}(-U_slot)
        BC_P[i] = HydroSolver.Dirichlet{Float64}(0)
    end
    if bottom_coflow[i]
        BC_U_x[i] = HydroSolver.Dirichlet{Float64}(0)
        BC_U_y[i] = HydroSolver.Dirichlet{Float64}(U_coflow)
        BC_P[i] = HydroSolver.Dirichlet{Float64}(zero(Float64))
    end
    if top_coflow[i]
        BC_U_x[i] = HydroSolver.Dirichlet{Float64}(0)
        BC_U_y[i] = HydroSolver.Dirichlet{Float64}(-U_coflow)
        BC_P[i] = HydroSolver.Dirichlet{Float64}(zero(Float64))
    end
    if right_border[i]
        BC_P[i] = HydroSolver.Dirichlet{Float64}(zero(Float64))
    end


end
```

```julia
function cells_that_have_edge_in_list(mesh,bool_edges)
    result = Vector{Bool}(undef,length(mesh.cells))
    for i in 1:length(mesh.cells)
        result[i] = any(bool_edges[mesh.cells[i]])
    end
    return result
end
```

```julia
DiffusP = HydroSolver.diffusion(mesh,BC_quantity=BC_P);
DiffusU_x = HydroSolver.diffusion(mesh,BC_quantity=BC_U_x);
DiffusU_y = HydroSolver.diffusion(mesh,BC_quantity=BC_U_y);
DiffusU = [DiffusU_x,DiffusU_y]
∇ = HydroSolver.gradient(mesh)
J = HydroSolver.interpolation(mesh,method=:Central)
RhieChow = HydroSolver.rhiechow(mesh, interp=J, grad=∇)
R = HydroSolver.split(RhieChow)
Rnorm = norm.(RhieChow)
Sum=HydroSolver.face_sum(mesh);
fV = HydroSolver.face_vecs(mesh);
```

```julia
function fractional_step(mesh, U, BC_U, BC_P; dt=1e-6,r=0.1, kwargs...)
    interp = HydroSolver.interpolation(mesh,U=U; kwargs...)
    Convec_x  = HydroSolver.advection(mesh, U, BC_velocity=BC_U, BC_quantity=BC_U_x, summation=Sum, interp=interp)
    Convec_y  = HydroSolver.advection(mesh, U, BC_velocity=BC_U, BC_quantity=BC_U_y, summation=Sum, interp=interp)
    u = HydroSolver.split(U)
    U_star = U - dt* HydroSolver.reunite([Convec_x*u[1], Convec_y*u[2]])
    u_star = HydroSolver.split(U_star)
    U_star_star = U_star + dt*ν* HydroSolver.reunite([DiffusU[i] * u_star[i] for i in 1:2])
    RHS = Sum*(fV*(interp-r*Rnorm)*U_star_star) .* ρ/dt
    P = DiffusP.A \ (RHS-DiffusP.B)
    GradP = ∇*P
    new_U = U_star_star - dt/ρ * GradP
    return new_U, P
end
```

```julia
U = [Point([0.0,0.0]) for i in 1:N_cells];
P = 1e-6*rand(N_cells)
dist = Vector{Float64}(undef,0)
```

```julia
obs = Observable(HydroSolver.col(P))
fig, ax, data = Makie.mesh(plot_vertices,plot_faces, color=obs,shading=false)
splot= streamplot!(ax,sobs,0..2,0..2, color=:white)
Colorbar(fig[1, 2], data)
ax.title = string("Pressure and velocity for iteration",1)
record(fig, "pressure.mp4", framerate = 60) do io
    global U
    global P
    global splot
    for j in 1001:2000
        U,P=fractional_step(mesh, U, [BC_U_x,BC_U_y], BC_P; dt=5e-6,r=-1e-2, method=:Upwind)
        if j%10==0
            g(x,y) = Point2{Float64}(HydroSolver.continuous_interpolation(mesh,U,Point([x,y])))
            delete!(ax,splot)
            obs[] = HydroSolver.col(P)
            splot=streamplot!(ax,g,0..2,0..2, colormap=:linear_grey_0_100_c0_n256)
            ax.title = string("Pressure and velocity for iteration ",j)
            notify(obs)
            recordframe!(io)
        end
    end
end
```

```julia
function SIMPLE_iter(mesh, U, BC_U, BC_P)
    interp = HydroSolver.interpolation(mesh,U=U; kwargs...)
    Convec_x  = HydroSolver.advection(mesh, U, BC_velocity=[BC_U_x,BC_U_y], BC_quantity=BC_U_x, summation=Sum, interp=interp)
    Convec_y  = HydroSolver.advection(mesh, U, BC_velocity=[BC_U_x,BC_U_y], BC_quantity=BC_U_y, summation=Sum, interp=interp)
    trM_x = Convec_x - ν * DiffusU_ax[1]
    trM_y = Convec_y - ν * DiffusU_ax[2]
    trM = [trM_x,trM_y]
    M_x = trM_x.A
    M_y = trM_y.A
    M = [M_x,M_y]
    A_x = Diagonal(M_x)
    A_y = Diagonal(M_y)
    A = [A_x,A_y]
    iA_x = inv(A_x)
    iA_y = inv(A_y)
    iA = [iA_x,iA_y]
    interp = HydroSolver.interpolate(mesh,U,HydroSolver.Interpolation{:Upwind})
    u = HydroSolver.split(U)
    for axis in 1:2
        H = (A[axis]-M[axis])*u[axis]
        m  = trM[axis].B
        LHS = (DiffusU_ax[axis].A*iA[axis])
        RHS = Vector(D_u*iA*(H-m-d_p))
        GradP[axis] .= LHS\RHS
        nU[axis] .= iA*H - iA*GradP[axis] # M\(GradP[axis]-m) #
    end
    G_P = [Point([ GradP[1][i] , GradP[2][i] ]) for i in 1:length(mesh.cells)]
    newU = [Point([ nU[1][i] , nU[2][i] ]) for i in 1:length(mesh.cells)]
    return G_P
end
```
